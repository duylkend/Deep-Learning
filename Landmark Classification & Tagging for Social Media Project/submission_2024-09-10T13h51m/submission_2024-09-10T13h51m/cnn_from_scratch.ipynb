<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ 



Define the Problem and Requirements
Objective: The goal was to implement a pre-defined VGG-16 architecture, which is widely used for image classification tasks, while potentially customizing the number of output classes for our specific dataset. The model also incorporates regularization through dropout to prevent overfitting.

Input: Images of size 224x224 pixels, which is standard for models like VGG-16.
Output: A tensor of shape [batch_size, num_classes] representing the classification scores for each class.
Steps to Design the CNN Architecture (VGG-16)
1. Use Predefined VGG-16 Convolutional Blocks
VGG-16 consists of 13 convolutional layers and 5 max-pooling layers. Each block of convolutional layers uses small 3x3 filters, maintaining consistency throughout the architecture to capture low-level to high-level features efficiently.

Conv Layers (Block 1-5):
Reasoning: The VGG-16 model progressively increases the number of filters as you move deeper into the network (from 64 to 512). This allows the network to capture simple patterns like edges in the initial layers and more complex structures in the deeper layers.

Structure: The architecture follows a repeated structure of stacking two or three convolutional layers followed by a max-pooling layer to reduce the spatial dimensions.

Conv1-2 (64 filters): Captures basic edges and textures.

Conv3-4 (128 filters): Extracts more complex patterns.

Conv5-6-7 (256 filters): Captures high-level features and object components.

Conv8-9-10 (512 filters): Extracts very detailed patterns relevant to image classification.

Conv11-12-13 (512 filters): Deepest layers extract abstract and complex patterns, preparing for classification.

2. Max-Pooling Layers
MaxPool2D (5 times):
Reasoning: Each max-pooling layer (with a 2x2 kernel and stride of 2) helps reduce the spatial dimensions, controlling the complexity and computational cost while preserving important features.
This downsampling helps the model focus on the most important features by progressively reducing the image size to 7x7 pixels before reaching the fully connected layers.
3. Fully Connected Layers
After flattening the final feature maps from the convolutional blocks, the features are passed through fully connected (FC) layers for classification.

Flatten:

Reasoning: The 512 feature maps of size 7x7 are flattened into a 1D vector to be used as input for the fully connected layers.
FC1 (4096 units):

Reasoning: The first fully connected layer uses a large number of neurons to combine the high-level features from the previous layers. It captures the complex relationships between the features.
FC2 (4096 units):

Reasoning: The second fully connected layer refines the features and prepares them for the final classification.
Output Layer (FC3) (num_classes):

Reasoning: The final layer uses the number of neurons equal to the number of classes for the dataset. It provides the classification scores for each class.
4. Regularization with Dropout
Dropout (applied before FC1 and FC2):
Reasoning: Dropout is applied during training to prevent overfitting. It randomly deactivates neurons in the fully connected layers, ensuring the model generalizes better to unseen data.
5. Activation Function
ReLU Activation:
Reasoning: The ReLU (Rectified Linear Unit) activation function is applied after each convolutional and fully connected layer (except the final layer). ReLU helps introduce non-linearity, allowing the network to learn complex patterns.
6. Implement the Forward Pass
In the forward pass, the input tensor is processed through each convolutional block, followed by max-pooling layers. After the convolutional feature extraction, the tensor is flattened and passed through the fully connected layers to produce the final classification scores.

Forward Pass:
Reasoning: The model sequentially applies convolution, pooling, flattening, and fully connected layers to transform the input image into class predictions. The use of dropout and activation functions ensures the model remains robust and efficient in learning.

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 64        # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2       # fraction of the training data to reserve for validation
num_epochs = 100        # number of epochs for training
num_classes = 50       # number of classes. Do not change this
dropout = 0.3          # dropout for our model
learning_rate = 0.25  # Learning rate for SGD (or Adam)
opt = 'sgd'            # optimizer. 'sgd' or 'adam'
weight_decay = 0.00000001  # regularization. Increase this to combat overfitting
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel

# get the data loaders using batch_size and valid_size defined in the previous
# cell
# HINT: do NOT copy/paste the values. Use the variables instead
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size)

# instance model MyModel with num_classes and drouput defined in the previous
# cell
model = MyModel(num_classes=num_classes, dropout=dropout)

# load the weights in 'checkpoints/best_val_loss.pt'
checkpoint_path = "best_val_loss.pt"
checkpoint = torch.load(checkpoint_path)
model.load_state_dict(checkpoint)

# Get the optimizer using get_optimizer and the model you just created, the learning rate,
# the optimizer and the weight decay specified in the previous cell
optimizer = get_optimizer(model=model, learning_rate=learning_rate, optimizer=opt, weight_decay=weight_decay)


# Get the loss using get_loss
loss = get_loss()

optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="best_val_loss.pt",
    interactive_tracking=True
)
````
