## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 2: Train, validation and test

Let's train our transfer learning model! Let's start defining the hyperparameters:

```python
batch_size = 64  # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2  # fraction of the training data to reserve for validation
num_epochs = 12  # number of epochs for training
num_classes = 50  # number of classes. Do not change this
learning_rate = 0.002   # Learning rate for SGD (or Adam)
opt = 'sgd'      # optimizer. 'sgd' or 'adam'
weight_decay = 0.0001 # regularization. Increase this to combat overfitting
```

```python
from src.data import get_data_loaders
from src.optimization import get_optimizer, get_loss
from src.train import optimize
from src.transfer import get_model_transfer_learning

# Get a model using get_model_transfer_learning. Use one of the names reported here:
# https://pytorch.org/vision/0.10/models.html
# For example, if you want to load ResNet 18, use "resnet18"
# NOTE: use the hyperparameters defined in the previous cell, do NOT copy/paste the
# values
model_transfer = get_model_transfer_learning(model_name="resnet18",n_classes =num_classes)

# train the model
data_loaders = get_data_loaders(batch_size=batch_size)
optimizer = get_optimizer(
    model_transfer,
    learning_rate=learning_rate,
    optimizer=opt,
    weight_decay=weight_decay,
)
loss = get_loss()

optimize(
    data_loaders,
    model_transfer,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="model_transfer.pt",
    interactive_tracking=True
)
```

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/>  __Answer:__ 
1. Choosing the Base Model: 
I opted for ResNet16 as the base CNN architecture. The reason is that ResNet, or Residual Networks, are designed to solve the vanishing gradient problem, which can occur when training deep neural networks. By using residual connections, the model ensures that gradients flow through the network during backpropagation, even in very deep layers. This guarantees a more stable training process and better performance, making it a strong candidate for image classification tasks.

2. Why ResNet16: 
ResNet16 has a moderate depth, which strikes a balance between complexity and efficiency. With more layers, deeper networks can capture more complex features and patterns in images. However, they can also be prone to overfitting, especially when data is limited. ResNet16 provides enough layers to extract important image details, without the risk of overcomplicating the model. This depth also allows for effective feature extraction and edge detection, which is critical for fine-tuning on the current image dataset.

3. Pretrained Weights: 
I initialized the model with pretrained weights, which were trained on a large dataset like ImageNet. The advantage here is that the model has already learned general patterns from "in-the-wild" images, which are similar to the dataset I am fine-tuning it on. This approach significantly reduces training time and increases the model’s accuracy, as the network doesn't need to start from scratch.

4. Fine-Tuning: 
After selecting the ResNet16 architecture, I fine-tuned the model on the specific dataset. This involved unfreezing the final layers and training them on the task-specific data. Fine-tuning helped the model adjust to the particular patterns in the target dataset, while retaining the general image features learned from the pretrained weights.

5. Hyperparameter Tuning: 
To optimize performance, I experimented with various hyperparameters:

6. Learning rate: 
I started with a smaller learning rate to ensure the pretrained weights weren’t overwritten too quickly during fine-tuning.
Batch size: I tested different batch sizes to balance training time and model performance.
Epochs: I monitored the model's performance on the validation set and adjusted the number of epochs to prevent overfitting.
Validation and Performance: After hyperparameter tuning, the model was tested on the validation set, achieving a performance over 60%, which was the minimum requirement. Through careful selection of the architecture, fine-tuning, and hyperparameter adjustments, I optimized the model, aiming for 80% or higher accuracy.

## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 4: Export using torchscript

Now, just like we did with our original model, we export the best fit model using torchscript so that it can be used in our application:

```python
from src.predictor import Predictor
from src.helpers import compute_mean_and_std

# First let's get the class names from our data loaders
class_names = data_loaders["train"].dataset.classes

# Then let's move the model_transfer to the CPU
# (we don't need GPU for inference)
model_transfer = model_transfer.cpu()
# Let's make sure we use the right weights by loading the
# best weights we have found during training
# NOTE: remember to use map_location='cpu' so the weights
# are loaded on the CPU (and not the GPU)
model_transfer.load_state_dict(
    torch.load("model_transfer.pt", map_location="cpu")
)

# Let's wrap our model using the predictor class
mean, std = compute_mean_and_std()
predictor = Predictor(model_transfer, class_names, mean, std).cpu()

# Export using torch.jit.script
scripted_predictor = torch.jit.script(predictor)
scripted_predictor.save("transfer_exported.pt")
```

```python
import torch
from src.predictor import predictor_test
from src.helpers import plot_confusion_matrix

model_reloaded = torch.jit.load("transfer_exported.pt")

pred, truth = predictor_test(data_loaders['test'], model_reloaded)

plot_confusion_matrix(pred, truth)
```
